Criando uma classe em python

```python
name = input('Digite seu nome:\n')
age = int(input('Digite sua idade: \n'))
descricao = input('Digite uma breve descrição sobre você:\n')

class Pessoa:
    def __init__(self, nome, idade, desc):
        self.nome = name
        self.idade = age
        self.desc = descricao

    def saudacao(self):
        return f"Olá, meu nome é {self.nome} e tenho {self.idade} anos.\n Sobre mim: {self.desc}"

# Obj pessoa
pessoa = Pessoa(nome= name, idade=age, desc=descricao)
print(pessoa.saudacao())
```
